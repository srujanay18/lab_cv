4. Use the reshaped X_test and y_test data to create a k-nearest neighbor classifier of digit. What is the optimal number of neighbors k? If you can't determine this, play around with different values of k for your classifier.

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

k_list = [1, 3, 5, 7, 9, 11]
results = []

for k in k_list:
    model = KNeighborsClassifier(n_neighbors=k)
    model.fit(X_train_flat, y_train)
    y_pred = model.predict(X_test_flat)
    acc = accuracy_score(y_test, y_pred)
    results.append((k, acc))

print("k vs accuracy:", results)
best_k, best_acc = max(results, key=lambda t: t[1])
print("Best k:", best_k, "Accuracy:", best_acc)
```

5. For the optimal number of neighbors, how well does your predictor perform on the test set? Report the accuracy, compute a confusion matrix, and explain your findings.

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix, accuracy_score
import matplotlib.pyplot as plt
import numpy as np

k_final = 3

final_model = KNeighborsClassifier(n_neighbors=k_final)
final_model.fit(X_train_flat, y_train)
y_pred = final_model.predict(X_test_flat)

acc = accuracy_score(y_test, y_pred)
print(f"Test accuracy (k={k_final}): {acc:.4f}")

cm = confusion_matrix(y_test, y_pred)
print("Confusion matrix (counts):\n", cm)

plt.figure()
plt.imshow(cm, interpolation='nearest')
plt.title(f'Confusion Matrix (k={k_final})')
plt.colorbar()
ticks = np.arange(10)
plt.xticks(ticks, ticks)
plt.yticks(ticks, ticks)
plt.xlabel('Predicted label')
plt.ylabel('True label')
plt.tight_layout()
plt.show()
```

For the optimal number of neighbors (3), the predictor performs pretty on the test set. The accuracy is 0.9705 and the confusion matrix is strongly diagonal.

6. For your confusion matrix, which mistakes are most likely? Do you find any interesting patterns?

```python
import numpy as np
from sklearn.metrics import confusion_matrix

cm_counts = cm
row_sums = cm_counts.sum(axis=1, keepdims=True)
cm_rates = cm_counts / row_sums

pairs = []
for true_d in range(10):
    for pred_d in range(10):
        if true_d == pred_d:
            continue
        count = int(cm_counts[true_d, pred_d])
        rate = float(cm_rates[true_d, pred_d])
        pairs.append((true_d, pred_d, count, rate))

# top 10 mistakes by count
pairs_sorted = sorted(pairs, key=lambda x: x[2], reverse=True)
print("Top 10 misclassifications (true -> predicted) by count:")
for t, p, cnt, r in pairs_sorted[:10]:
    print(f"{t} -> {p}: {cnt} errors ({r:.2%} of true {t})")

# most common mistake per true digit
print("\nMost common mistake for each true digit:")
for t in range(10):
    row = [(p, int(cm_counts[t, p]), float(cm_rates[t, p])) for p in range(10) if p != t]
    p_best, cnt_best, rate_best = max(row, key=lambda x: x[1])
    print(f"{t}: predicted as {p_best}  {cnt_best} errors ({rate_best:.2%})")
```

The mistakes that are the most likely are for digit 7, which are predicted as 1, and are shown in the confusion 21 times/errors. A pattern that is interesting are that within the top 10 misclassifications, most of them are wrongly predicted as either 0, 1, 3, 5, 7, and 9; which are all odd numbers besides 0. Additionally, all the top 10 misclassification percentages are less than around 2% within their true class.
